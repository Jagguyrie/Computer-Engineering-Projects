```python
import pandas as pd

df = pd.read_csv("titanic.csv")

df = df[~df["Age"].isnull()]
df = df[~df["Fare"].isnull()]
df = df[~df["Parch"].isnull()]
df = df[~df["SibSp"].isnull()]

df["Age"] = df["Age"].astype("int64")
df["Fare"] = df["Fare"].astype("int64")

#Q7
c = 3

q1a , q3a = df["Age"].quantile([0.25, 0.75])
iqr_a = q3a - q1a
tmina = q1a - (c * iqr_a)
tmaxa = q3a + (c * iqr_a)
print(tmina, tmaxa)
q1f, q3f = df["Fare"].quantile([0.25, 0.75])
iqr_f = q3f - q1f
tminf = q1f - (c * iqr_f)
tmaxf = q3f + (c * iqr_f)
print(tminf, tmaxf)
q1p, q3p = df["Parch"].quantile([0.25, 0.75])
iqr_p = q3p - q1p
tminp = q1p - (c * iqr_p)
tmaxp = q3p + (c * iqr_p)
print(tminp, tmaxp)
q1s, q3s = df["SibSp"].quantile([0.25, 0.75])
iqr_s = q3s - q1s
tmins = q1s - (c * iqr_s)
tmaxs = q3s + (c * iqr_s)
print(tmins, tmaxs)

df_no = df[(df["Age"] >= tmina)  (df["Age"] <= tmaxa)]
df_no = df[(df["Fare"] >= tminf) and (df["Fare"] <= tmaxf)]
df_no = df[(df["Parch"] >= tminp) and (df["Parch"] <= tmaxp)]
df_no = df[(df["SibSp"] >= tmins) and (df["SibSp"] <= tmaxs)]

df.hist(column = "Age")
df.hist(column = "Fare")
df.hist(column = "Parch")
df.hist(column = "SibSp")
```
