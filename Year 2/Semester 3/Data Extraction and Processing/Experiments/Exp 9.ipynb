```python
import pandas as pd
df = pd.read_csv('indian_food.csv')
df1 = df

#Q1
print('Size of dataframe is', df.size, '\n')
#Q2 
print('Shape of the dataframe is', df.shape, '\n')
#Q3
print('Data types of individual columns are:\n', df.dtypes, '\n')
#Q4
print((f'There are {len(df["name"])} dishes in indian cuisine\n'))
#Q5
print('Each indian dish has a name, ingredients required, diet type, preparation time, flavor, course type, origin state, and region.\nIn total there are 8 features.\n')
print('Statistics about cook time are:\n', df["cook_time"].describe())
print('Statistics about preparation time\n', df["prep_time"].describe())
#Q6
#Yes, missing values exist
for i in range(0, len(df)):
    if df['prep_time'][i] == -1:
        df1['prep_time'][i] = 'NaN'
    if df['cook_time'][i] == -1:
        df1['cook_time'][i] = 'NaN'
    if df['flavor_profile'][i] == '-1':
        df1['flavor_profile'][i] = 'NaN'
    if df['course'][i] == '-1':
        df1['course'][i] = 'NaN'
    if df['state'][i] == '-1':
        df1['state'][i] = 'NaN'
    if df['region'][i] == '-1':
        df1['region'][i] = 'NaN'
print(df1)
```
