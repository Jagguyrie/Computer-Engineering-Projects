```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("laptop_o.csv")
df_e = pd.read_csv("laptop_edited.csv")
#print("The original dataframe is:\n",df)
'''
print("Summary of the dataframe is:")
print(df.info())
print(df_e.describe())
'''
#Finding number of NA values for each column
missing = {}
for i in df.columns:
    missing[i] = df[i].isna().sum()
print("The missing values are: ")
for i,j in missing.items():
    print(f'{i} = {j}')

#Handling NA values with fillna
newdf = df.fillna(-1)
print(newdf)

#Replacing all NA values with -1
print(newdf.info())

#Finding mean of rating, ignoring -1
mean,tot_ignore = 0,0
for i in range(len(newdf)):
    if df['screen_size (inches)'][i] == -1:
        tot_ignore += 1
        continue
    else:
        mean += newdf['screen_size (inches)'][i]
mean = mean/(len(newdf) - tot_ignore)
print("The mean of screen size is: ", mean)

#Finding count of laptops which have ratings greater than 4 and price less than 2000
count1 = 0
for i in range(len(newdf)):
    if df['rating'][i] >= 4 and (df['price'][i] <= 2000 and df['price'][i] > 0):
        count1 += 1
print("Total laptops which have ratings greater than 4 and price less than 2000 are: ",count1)

#Adding a new column where based on ratings, 1-2.5: Bad, 2.5-4: Good, 4-5: Great
rate = []
for i in range(len(newdf)):
    if newdf['rating'][i] >= 1 and newdf['rating'][i] < 2.5:
        rate.append('Bad')
    elif newdf['rating'][i] >= 2.5 and newdf['rating'][i] < 4:
        rate.append('Good')
    elif newdf['rating'][i] >= 4 and newdf['rating'][i] <= 5:
        rate.append('Great')
    else:
        rate.append("No rating")
newdf['rate'] = rate
print("On adding new column, dataframe is:\n",newdf)

#Finding first 20 elements of dataframe
print("The first 20 elements are:\n",newdf.head(20))

#Making histograms
plt.hist(newdf['rating'],range = (1,5))
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Histogram of rating")
plt.show()

plt.hist(newdf['screen_size (inches)'],range = (10,18))
plt.xlabel("Screen size")
plt.ylabel("Frequency")
plt.title("Histogram of screen size")
plt.show()

#Making a pie chart for rate
count = [0,0,0]
for i in range(len(newdf)):
    if newdf['rate'][i] == "Bad":
        count[0] += 1
    elif newdf['rate'][i] == "Good":
        count[1] += 1
    elif newdf['rate'][i] == "Great":
        count[2] += 1
plt.pie(count, labels = ["Bad","Good","Great"])
plt.show()

#Finding number of Dell,Lenovo,HP and Acer laptops
company = ["Dell","Lenovo","HP","Acer"]
tot = [0,0,0,0]
for i in range(len(newdf)):
    if newdf['brand'][i].upper() == "DELL":
        tot[0] += 1
    elif newdf['brand'][i].upper() == "LENOVO":
        tot[1] += 1
    elif newdf['brand'][i].upper() == "HP":
        tot[2] += 1
    elif newdf['brand'][i].upper() == "ACER":
        tot[3] += 1
print(f"The count for each company: \n{company[0]}: {tot[0]}\n{company[1]}: {tot[1]}\n{company[2]}: {tot[2]}\n{company[3]}: {tot[3]}")

#Finding number of unique values for each column
for i in newdf.columns:
    st = set(newdf[i])
    print(f"The number of unique values for {i} is: {len(st)}")
```
